```python
# Cell 1: Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.utils.data_science import cosine_similarity, softmax, normalize_embeddings

# Cell 2: Create sample data
embeddings = np.random.randn(1000, 768)
texts = [f"Sample text {i}" for i in range(1000)]
scores = np.random.rand(1000)

# Cell 3: Analyze embeddings
normalized = normalize_embeddings(embeddings)
sample_a, sample_b = normalized[0], normalized[1]
similarity = cosine_similarity(sample_a, sample_b)
print(f"Cosine similarity: {similarity:.3f}")

# Cell 4: Create DataFrame
df = pd.DataFrame({
    "text": texts,
    "embedding": list(embeddings),
    "score": scores
})
print(df.head())
print(df.describe())
```
